```python
import pandas as pd

df = pd.read_csv('C:/Users/user/OneDrive/Desktop/books_toscrape.csv')
df.head()
```

```python
# DataFrame structure and data types
print(df.info())

# Statistical summary (for numerical columns)
print(df.describe())

# Display the first 5 rows
print(df.head())
```

```python
df['Price'] = df['Price'].str.replace('£', '').astype(float)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the size of the plot
plt.figure(figsize=(10,6))

# Plot a histogram of the 'Price' column with KDE curve (Kernel Density Estimate)
sns.histplot(df['Price'], bins=20, kde=True)

# Title of the plot
plt.title('Distribution of Book Prices')

# Label for the X-axis
plt.xlabel('Price (£)')

# Label for the Y-axis
plt.ylabel('Number of Books')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size (width=12, height=6)
plt.figure(figsize=(12,6))

# Create a horizontal bar chart showing count of books per category
# The categories are ordered by their counts (most frequent first)
sns.countplot(y='Category', data=df, order=df['Category'].value_counts().index)

# Title of the chart
plt.title('Number of Books by Category')

# Label for the X-axis
plt.xlabel('Number of Books')

# Label for the Y-axis
plt.ylabel('Category')

# Display the plot
plt.show()
```

```python
rating_map = {'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}
df['Rating_num'] = df['Rating'].map(rating_map)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size (width=8, height=5)
plt.figure(figsize=(8,5))

# Create a scatter plot to show relationship between rating (x) and price (y)
sns.scatterplot(x='Rating_num', y='Price', data=df)

# Title of the plot
plt.title('Relationship between Price and Rating')

# Label for X-axis (Rating)
plt.xlabel('Rating')

# Label for Y-axis (Price in £)
plt.ylabel('Price (£)')

# Show the plot
plt.show()
```
